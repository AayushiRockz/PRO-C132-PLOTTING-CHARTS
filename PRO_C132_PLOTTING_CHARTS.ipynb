```python
#importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the file as df
df = pd.read_csv("star_with_gravity.csv")

#Creating lists of radius , mass and gravity
radii_list = df["Radius"].to_list()
masses_list = df["Mass"].to_list()
gravity_list = df["Gravity"].to_list()

#sorting the lists (in increasing order)
radii_list.sort()
masses_list.sort()
gravity_list.sort()


```

Plotting Graphs 

```python
#plotting a line graph of Radius and mass
plt.plot(radii_list,masses_list)

plt.title("Radius & Mass of the Star (line graph)")
plt.xlabel("Radius")
plt.ylabel("Mass")

plt.show()
```

```python
#plotting a scatter plot between radius and mass
plt.scatter(radii_list,masses_list)
plt.title("Scatter plot of Radius and Mass")
plt.xlabel("Radius")
plt.ylabel("Mass")
plt.show()
```

```python
#plotting a line graph of mass and gravity
plt.plot(masses_list,gravity_list)

plt.title("Mass and gravity of the Star")
plt.xlabel("Mass")
plt.ylabel("Gravity")

plt.show()
```
